```python
#Cleaning all with corr > 0.85
file = file.drop(['Crossing', 'Finishing', 'Short Passing', 'Volleys', 'Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control', 'Acceleration', 'Sprint Speed', 'Agility', 'Balance', 'Marking', 'Standing Tackle', 'Sliding Tackle', 'GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes'], axis=1)
fileVal = fileVal.drop(['Crossing', 'Finishing', 'Short Passing', 'Volleys', 'Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control', 'Acceleration', 'Sprint Speed', 'Agility', 'Balance', 'Marking', 'Standing Tackle', 'Sliding Tackle', 'GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes'], axis=1)
```

```python
file_num = file.select_dtypes('number')
```

```python
#I'm going to replace the NaN with the mean
for col in file_num:
        mean = int(file_num[col].mean())
        file[col] = file_num[col].fillna(mean)
```

```python
file_num = fileVal.select_dtypes('number')
for col in file_num:
        mean = int(file_num[col].mean())
        fileVal[col] = file_num[col].fillna(mean)
```

```python
#Checking again corr from all with all (cleaned all with corr>85%)
file = file.drop(['Total Stats', 'Skill', 'Mentality', 'Long Shots', 'Positioning', 'Defending', 'Interceptions', 'Strength'], axis=1)
fileVal = fileVal.drop(['Total Stats', 'Skill', 'Mentality', 'Long Shots', 'Positioning', 'Defending', 'Interceptions', 'Strength'], axis=1)
```

```python
file_num = file.select_dtypes('number')
```

```python
file_cat = file.select_dtypes('object')
```

```python
#Filling NaN in Cat columns
```

```python
file['A/W'] = file['A/W'].fillna('Medium')
fileVal['A/W'] = fileVal['A/W'].fillna('Medium')
```

```python
file['D/W'] = file['D/W'].fillna('Medium')
fileVal['D/W'] = fileVal['D/W'].fillna('Medium')
```

```python
#I'm going to eliminate the symbols or in those columns which have it
file['W/F'] = file['W/F'].str.rstrip('★').astype('int')
file['SM'] = file['SM'].str.rstrip('★').astype('int')
file['IR'] = file['IR'].str.rstrip('★').astype('int')
file['Weight'] = file['Weight'].str.rstrip('lbs').astype('int')
```

```python
fileVal['W/F'] = fileVal['W/F'].str.rstrip('★').astype('int')
fileVal['SM'] = fileVal['SM'].str.rstrip('★').astype('int')
fileVal['IR'] = fileVal['IR'].str.rstrip('★').astype('int')
fileVal['Weight'] = fileVal['Weight'].str.rstrip('lbs').astype('int')
```

```python
#Converting of height to integer
def height (x):
    x=x.replace('"','')
    f,i=x.split("\'")
    x=round(int(f)+int(i)/12,2)
    return x
file['Height']=list(map(height, file['Height']))
fileVal['Height']=list(map(height, fileVal['Height']))
```

```python
# convertion of amounts (€) to integer
def amount (x):
    x=x.replace('€','')
    if 'K' in x:
        scalar=1000
        x=x.replace('K','')
    elif 'M' in x:
        scalar=1000000
        x=x.replace('M','')
    else:
        scalar=1
    x=pd.to_numeric(x, errors='coerce')*scalar
    return x
file['Value']=list(map(amount, file['Value']))
file['Wage']=list(map(amount, file['Wage']))
file['Release Clause']=list(map(amount, file['Release Clause']))
```

```python
fileVal['Value']=list(map(amount, fileVal['Value']))
fileVal['Wage']=list(map(amount, fileVal['Wage']))
fileVal['Release Clause']=list(map(amount, fileVal['Release Clause']))
```

```python
def hits (x):
    if 'K' in x:
        scalar=1000
        x=x.replace('K','')
    else:
        scalar=1
    x=pd.to_numeric(x, errors='coerce')*scalar
    return x
file['Hits']=list(map(hits, file['Hits']))
fileVal['Hits']=list(map(hits, fileVal['Hits']))
```

```python
#Again, I'll eliminate categories with corr>85%
file = file.drop(['Wage', 'Release Clause'], axis=1)
fileVal = fileVal.drop(['Wage', 'Release Clause'], axis=1)
```

```python
file = file.drop(['LS', 'RS', 'LF', 'CF', 'RF', 'LAM', 'RAM', 'LM', 'LCM', 'RCM', 'RM', 'LWB', 'LDM', 'RDM', 'RWB', 'LCB', 'RCB'], axis=1)
fileVal = fileVal.drop(['LS', 'RS', 'LF', 'CF', 'RF', 'LAM', 'RAM', 'LM', 'LCM', 'RCM', 'RM', 'LWB', 'LDM', 'RDM', 'RWB', 'LCB', 'RCB'], axis=1)
```

```python
def progression (x):
    f,i=x.split("+")
    x=int(i)
    return x
file['GK']=list(map(progression, file['GK']))
file['CB']=list(map(progression, file['CB']))
file['LB']=list(map(progression, file['LB']))
file['RB']=list(map(progression, file['RB']))
file['CM']=list(map(progression, file['CM']))
file['CAM']=list(map(progression, file['CAM']))
file['CDM']=list(map(progression, file['CDM']))
file['ST']=list(map(progression, file['ST']))
file['RW']=list(map(progression, file['RW']))
file['LW']=list(map(progression, file['LW']))
```

```python
fileVal['GK']=list(map(progression, fileVal['GK']))
fileVal['CB']=list(map(progression, fileVal['CB']))
fileVal['LB']=list(map(progression, fileVal['LB']))
fileVal['RB']=list(map(progression, fileVal['RB']))
fileVal['CM']=list(map(progression, fileVal['CM']))
fileVal['CAM']=list(map(progression, fileVal['CAM']))
fileVal['CDM']=list(map(progression, fileVal['CDM']))
fileVal['ST']=list(map(progression, fileVal['ST']))
fileVal['RW']=list(map(progression, fileVal['RW']))
fileVal['LW']=list(map(progression, fileVal['LW']))
```

```python
#One hot encoding
```

```python
from sklearn.preprocessing import OneHotEncoder

y_train = file['OVA']
X = file.drop(['OVA'], axis=1)
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = object)
```

```python
encoder = OneHotEncoder(drop='first').fit(X_cat)
cols = encoder.get_feature_names(input_features=X_cat.columns)
X_cat_encode = pd.DataFrame(encoder.transform(X_cat).toarray(),columns=cols)
X_cat_encode.head()
```

```python
X_train = pd.concat([X_num, X_cat_encode], axis=1)
```

```python
y_test = fileVal['OVA']
X = fileVal.drop(['OVA'], axis=1)
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = object)
```

```python
cols = encoder.get_feature_names(input_features=X_cat.columns)
X_cat_encode = pd.DataFrame(encoder.transform(X_cat).toarray(),columns=cols)
```

```python
X_test = pd.concat([X_num, X_cat_encode], axis=1)
```

```python
#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)
```

```python
from sklearn import linear_model
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
#Now I'll calculate R2 score, MSE, RMSE and MAE
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
train_pred = lm.predict(X_train)
test_pred = lm.predict(X_test)
r2t = r2_score(y_train, train_pred)
print("R2 for train:", r2t)
r2s = r2_score(y_test, test_pred)
print("\nR2 for score:", r2s)
train_mse = mean_squared_error(train_pred, y_train)
print ('\nTrain MSE: {}'.format(train_mse))
test_mse = mean_squared_error(test_pred, y_test)
print ('\nTest MSE: {}'.format(test_mse))
train_rmse = math.sqrt(train_mse)
print("\nTrain RMSE:", train_rmse)
test_rmse = math.sqrt(test_mse)
print("\nTest RMSE:", test_rmse)
train_mae = mean_absolute_error(y_train, train_pred)
print("\nTrain MAE:", train_mae)
test_mae = mean_absolute_error(y_test, test_pred)
print("\nTest MAE:", test_mae)
```
